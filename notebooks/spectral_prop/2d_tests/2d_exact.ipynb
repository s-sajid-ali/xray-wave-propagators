```python
import numpy as np
import matplotlib.pyplot as plt
from xwp.exact_2d import exact_prop
```

```python
wavel = 0.5*10**(-6)
pi = np.pi
z = 100000
N = 100
L_in  = 5e-1

in_wave = np.zeros((N,N))
in_wave[int(N/2)-int(N/8):int(N/2)+int(N/8),int(N/2)-int(N/8):int(N/2)+int(N/8)] = 1
out_wave = np.zeros((100,100),dtype='complex128')
```

```python
exact_prop(in_wave,out_wave,L_in,L_in,wavel,z)
```

```python
f, (ax1,ax2) = plt.subplots(1,2)
ax1.imshow(np.abs(in_wave),cmap='jet')
ax2.imshow(np.abs(out_wave),cmap='jet')
plt.show()
```

```python
%timeit exact_prop(in_wave,out_wave,L_in,L_in,wavel,z)
```
